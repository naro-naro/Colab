# 절대 경로

# 상대 경로
- 상대 경로는 현재 위치한 곳을 기준으로 하는 경로
- /는 루트
- ./는 현재 위치
- ../는 현재 위치의 상위 폴더
- ../../는 두 단계 위의 상위 폴더
