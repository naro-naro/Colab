```python
# 불러오기
import pandas as pd
```

# DataFrame
- 2차원의 자료구조

```python
df = pd.DataFrame(
    {
        'a' : [4,5,6],
        'b' : [7,8,9],
        'c' : [10,11,12]
    },
    index = [1,2,3]
)

df
```

# Series
- 1차원의 자료구조

```python
df['a']
```

```python
df[['a']]
```

# Subset

```python
# boolean 값으로 결과 출력
df["a"] > 4
```

```python
df[df["a"] > 4]
```

```python
df['a', 'b']
```

```python
# 둘 이상의 컬럼 선택 ; 리스트로 감싸기
df[['a', 'b']]
```

# Summarize Data

```python
# 원소의 빈도수
df['a'].value_counts()
```

```python
len(df)
```

# Reshaping
## sort_values, drop

```python
# a컬럼 기준 요소값 정렬
df['a'].sort_values()
```

```python
# 전체 컬럼 - a컬럼 기준으로 요소값 정렬, 역순으로
df.sort_values("a", ascending=False)
```

```python
# 특정 컬럼(c) 삭제
df = df.drop(["c"], axis=1)
df
```

# Group Data
## Groupby, pivot_table

```python
df.groupby(['a'])['b'].mean()
```

```python
df.groupby(['a'])['b'].agg(['mean', 'sum', 'count'])
```

```python
df.groupby(['a'])['b'].describe()
```

```python
pd.pivot_table(df, index='a')
```

```python
pd.pivot_table(df, index='a', values='b', aggfunc='sum')
```
