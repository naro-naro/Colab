- 문항별 응답내역이 'question'에 담겨 있을 때, **조건문과 반복문을 사용하여** 아래와 같은 결과가 출력되도록 코드를 작성해보세요.

````python
# 응답 설문 문항
question = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6_1', 'Q6_2', 'Q6_3', 'Q6_4', 'Q6_5', 'Q6_6',
       'Q6_7', 'Q6_8', 'Q6_9', 'Q6_10', 'Q6_11', 'Q6_12', 'Q7_1', 'Q7_2',
       'Q7_3', 'Q7_4', 'Q7_5', 'Q7_6', 'Q7_7', 'Q8', 'Q9', 'Q10_1', 'Q10_2',
       'Q10_3']
````

````python
# single choice 문항만 담을 리스트 생성
single_choice = []

# 복수 응답 문항이 아닌지 "_" 포함 여부로 판단하여
# 아니라면 해당 문항 번호를 single_choice 리스트에 담기
for i in question:
    if "_" not in i:
        single_choice.append(i)

# single choice 리스트 출력
print(single_choice)
````

- 갭마인더 사이트에서는 연도별, 국가별 GDP와 기대수명 데이터를 제공하고 있는데, 대표적으로 파이썬 라이브러리 중 'seaborn'에서 제공되는 예제 데이터가 있습니다. 오늘은 이 데이터를 활용해 문제를 풀어보겠습니다.
- 다음의 데이터는 연도, 국가별 기대수명을 나타내고 있는 데이터 입니다. ***2011년 부터의 연도별, 국가별 평균 기대수명***을 구해주세요. (2011년도 포함되게 구합니다.)


- [필수 조건]
    - **groupby** 나 **pivot_table**을 활용합니다.
    - groupby 로 구할 때 **unstack()**이라는 기능을 사용하여 아래와 같이 컬럼에 인덱스 값을 올려서 표기할 수 있습니다.
    - 만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터의 결과값이 아래와 같다면, 출력형태는 조금 달라도 괜찮습니다.

````python
# 연도, 국가별 기대수명 데이터 불러오기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
````

## groupby

````python
# 연도별 국가별 평균 기대수명을 life_expectancy 로 저장
life_expectancy = df.groupby(['Year', "Country"])['Life_Expectancy'].mean().unstack()
life_expectancy
````

````python
# 연도별 국가별 평균 기대수명 life_expectancy 을
# 2011년부터의 데이터만 저장
life_expectancy_2011 = life_expectancy[life_expectancy.index >= 2011]
life_expectancy_2011
````

## pivot_table

````python
# 전체 기간의 연도별 국가별 평균 기대수명을 life_expectancy2 에 저장
life_expectancy2 = df.pivot_table(index="Year", columns='Country', values = 'Life_Expectancy')
life_expectancy2
````

````python
# life_expectancy2 중 2011 이후의 데이터만 출력
life_expectancy2_2011 = life_expectancy2.loc[2011:]
life_expectancy2_2011
````
