```python
import matplotlib.pyplot as plt
import pandas as pd
filename = '/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/iris.csv'
iris = pd.read_csv(filename)
iris.head()
```

```python
# 막대 그래프 그리기
iris.SepalLength[:10].plot(kind='bar', rot=30) # rot= 글자 기울기

plt.rc('font', family='NanumBarunGothic')
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
# 품종 별 평균값 계산
df2 = iris.groupby(iris["Species"]).mean()
df2
```

```python
# 수평 막대 그래프 그리기
df2.SepalLength[:].plot(kind='barh', rot=0)

plt.rc('font', family='NanumBarunGothic')
plt.title("품종별 SelpalLength 그래프")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이의 평균값")
plt.show()
```

```python
# 히스토그램 그리기
iris.plot(kind='hist')
plt.title("각 특징 값들의 빈도수 히스토그램")
plt.show()
```

```python
# 박스플롯 그리기
# 특이치 관측에 유리
iris.plot(kind='box')
plt.title("각 특징 값들의 빈도수에 대한 박스 플롯")
plt.xlabel("특징")
plt.ylabel("데이터 값")
plt.show()
```

# 타이타닉 데이터 분석

```python
import pandas as pd
filename = '/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/train.csv'
titanic = pd.read_csv(filename)
titanic.head()
```

* **데이터 확인**

```python
# 기본 정보
titanic.info()
```

```python
# 기초 통계량
titanic.describe()
```

```python
# 요금 순 내림차순 정렬
titanic.sort_values("Fare",ascending=False)
```

```python
# 생조자 별 인원수 구하기
pd.DataFrame(titanic['Survived'].value_counts())
```

* **데이터 전처리**

```python
# 결측치 확인*********
titanic.isnull().sum()
```

```python
# 객실 번호 칼럼 삭제
titanic.drop(["Cabin"], axis=1, inplace=True)
titanic.columns
```

```python
# 결측 데이터 최빈값으로 대체하기(1)
# 최빈값 찾기
titanic['Embarked'].value_counts()
```

```python
# 결측 데이터 최빈값으로 대체하기(2)
# 최빈값 대체하기
titanic['Embarked'] = titanic['Embarked'].fillna("S")
titanic.isnull().sum()
```

```python
# 결측 데이터 평균값으로 대체하기
avg = titanic['Age'].mean()
titanic["Age"] = titanic["Age"].fillna(avg)
titanic.isnull().sum()
```

* **데이터 그룹핑**

```python
# 항구별 Pcalss컬럼의 평균값 출력
titanic["Pclass"].groupby(titanic["Embarked"]).mean()
```

```python
# Pcalss, Sex 칼럼을 기준으로 평균 구하기
titanic.groupby(["Pclass", "Sex"]).mean() # 오류날 때 : mean(numeric_only=True)
```

* **데이터 시각화**

```python
# 막대 그래프 그리기

import seaborn as sns
sns.countplot('Pclass', hue='Survived', data=titanic)
```

```python
# 히트맵 그래프 그리기
sns.heatmap(data=titanic.corr(), annot=True, cmap='Blues')
```

```python
# 히트맵 그래프 그리기
# annot=False - 칸마다 숫자 표시
# cmap='Reds' - 그래프 색깔
sns.heatmap(data=titanic.corr(), annot=False, cmap='Reds')
```
