```python
filename = '/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/csv/fine_dust.xlsx'
data = pd.read_excel(filename)
data2019 = data[2019] # 2019년 데이터만 data2019에 저장
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,4)) # 그래프 크기 지정
plt.plot(data2019, color='b', marker='o') # 원형 마커 지정
plt.xlabel('area') # x축 이름 지정
plt.ylabel('micrometer') # y축 이름 지정
plt.title('2019 Fine Dust Line Graph') # 그래프 제목 지정
plt.grid() # 격자 활성화
plt.show() # 그래프 출력
```

```python
plt.figure(figsize=(15,4))

for year in range (2016,2020):
    chartdata = data[year]
    plt.plot(chartdata, marker='o', label=year)

plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2001~2019 Fine Dust Line Graph')
plt.legend()
plt.grid()
plt.show()
```

# 자료구조와 알고리즘

알고리즘
1. 입력 숫자들을 저장한다.
2. 가장 큰 숫자에 0을 저장한다.
3. 숫자를 읽어와 저장한다.
4. 저장된 숫자와 가장 큰 숫자를 비교하여
큰 숫자를 다시 가장 큰 숫자로 저장한다.
5. 입력 숫자의 개수만큼 3, 4 를 반복한다.
6. 가장 큰 숫자를 출력한다.

```python
# 가장 큰 숫자 찾기

# 입력 - 자료구조
num =[37,29,69,30,15,55,90,87,73,20]
largist_num=0

# 가장 큰 숫자 찾기 - 문제해결 알고리즘
for i in num:
    if largist_num<i: # 0<37(T), 37<29(F),
        largist_num = i # 37,            ,

# 출력
print(largist_num)
```
