# 판다스 데이터 선택하기

```python
import pandas as pd
weather = '/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/weather.csv'
df = pd.read_csv(weather)
df.head(3)
```

* 열 선택

```python
# 하나의 열 선택
df['date']
```

```python
type(df['max_wind'])
# 열 하나의 데이터 타입은 시리즈
```

```python
# 여러 열 선택 : [['col1' , 'col2']]
df[['date', 'max_wind']]
```

```python
# 여러 열 선택 : loc()
df.loc[:,['date', 'temp']] # : 전체범위
# 행은 전체범위인데 열은 선택한것만
```

```python
# 여러 열 선택 : loc()
df.loc[0:3,['date', 'max_wind']] # 0~3행까지만 나오게
```

```python
# 여러 열 선택 : loc()
df.loc[0:3:2,['date', 'max_wind']] # 0~3행까지만 나오게, 2의 간격으로
```

```python
# 여러 열 선택 : loc()
df.loc[0:365:30,['date', 'max_wind']] # 1년간 30일 간격으로
```

```python
# 여러 열 선택 : iloc()
df.iloc[:, [0,1]]
```

```python
# 여러 열 선택 : iloc()
df.iloc[0:365:30, [0,2]] # 0번째, 2번째 열 출력
```

* 행 선택하기

```python
df[0:3] # 0~2행
```

```python
df.iloc[2] # 2행만 인덱스로 선택
```

```python
df.index=df['date'] # date를 인덱스화
print(df)
```

```python
# 특정날짜 인덱스에 해당하는 temp와 mean_wind 값 출력
df.index=df['date'] # date를 인덱스화
df.loc['2010-08-01',['temp','mean_wind']] # 해당 이름의 인덱스의 온도와 평균풍속 출력
```

```python
df.index=df['date'] # date를 인덱스화
df.iloc[0:3,[1,3]] # 해당 이름의 인덱스의 온도와 평균풍속 출력
```

```python
df.index=df['date'] # date를 인덱스화
df.loc['2010-08-01':'2019-08-01':365,['temp','mean_wind']]
# 100801과 190801 사이의 인덱스를 365간격으로 온도와 평균풍속 칼럼 출력
```

* 불 인덱싱

```python
w=  df['temp'] >= 30 # 온도가 30이상인 데이터 출력
df[w]
```

```python
# 가장 더웠던 날
w =  df['temp'] == df['temp'].max()  # 비교해서 최댓값이면 w에 저장
df[w]
```

```python
# 조건이 두개 이상
# | or
# & and
w=  (df['temp'] >= 30) & (df['max_wind']>=9)
df[w]
```
