```python
# 사용
import matplotlib.pyplot as plt
```

# 숫자 입력
- **숫자 리스트 1개 입력**
  - y값으로 인식
  - x값은 [0, 1, 2, 3]으로 디폴트 설정됨
  - 튜플, 넘파이 배열 형태도 가능
  - **plt.show()** : 그래프를 화면에 표시
- **숫자 리스트 2개 입력**
  - 순서대로 x, y값으로 인식
  - 순서쌍 (x, y)로 매칭된 값을 좌표평면 위에 그래프 시각화

```python
# 1개 숫자 리스트 입력
plt.plot([2,3,4,5])
plt.show()
```

```python
# 2개 리스트 입력
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

# 축 레이블 설정
- **xlabel()** : x축 레이블 설정
- **ylabel()** : y축 레이블 설정

```python
# xlabel() : x축 레이블 설정
# ylabel() : y축 레이블 설정
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

# 범례(Legend) 설정
- **범례** : 그래프에 데이터의 종류를 표시하기 위한 텍스트
- **legend()** : 그래프에 범례 표시
- **plot()** : label 파라미터값으로 삽입

```python
# legend() : 범례 표시
plt.plot([1,2,3,4], [1,4,9,16], label="Square")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.legend()
plt.show()
```

# 축 범위 설정
- **xlim([xmin, xmax])** : x축이 표시되는 범위 지정
- **ylim([ymin, ymax])** : y축이 표시되는 범위 지정
- **axis([xmin, xmax, ymin, ymax])** : x,y축이 표시되는 범위 지정
- 입력 값이 없으면, 자동으로 데이터에 맞게 범위 지정

```python
# xlim([xmin, xmax]) : x축이 표시되는 범위 지정
# ylim([ymin, ymax]) : y축이 표시되는 범위 지정
plt.plot([1,2,3,4], [3,6,9,12], label="y = 3x")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.xlim([0,5])
plt.ylim([0,15])
plt.legend()
plt.show()
```

```python
# axis([xmin, xmax, ymin, ymax]) : x,y축이 표시되는 범위 지정
plt.plot([1,2,3,4], [1,4,9,16], label="y = x^2")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.axis([0,6,0,20])
plt.legend()
plt.show()
```

# 선 종류 설정
- **plot()함수의 포맷문자열**
  - **-** : solid
  - **--** : dashed
  - **-.** : dotted
  - **:** : dashdotted
- **plot()함수의 linestyle** 파라미터 값으로 삽입
  - linestyle = 'solid'
  - linestyle = 'dashed'
  - linestyle = 'dotted'
  - linestyle = 'dashdotted'
- 튜플을 사용해 선의 종류 커스터마이징
  - (0, (1, 1)) [dotted]
  - (0, (5, 5)) [dashed]
  - (0, (3, 5, 1, 5)) [dashdotted]

```python
# plot() 함수의 포맷 문자열 사용
plt.plot([1,2,3,4], [0.5, 1, 1.5, 2], '-', color='C0', label = "y = 1/2x")
plt.plot([1,2,3,4], [1,2,3,4], '--', color='red', label="y = x")

# plot() 함수의 linestyle 값 사용
plt.plot([1,2,3,4], [2,4,6,8], linestyle='dotted', color='black', label="y = 2x")
plt.plot([1,2,3,4], [3,3,3,3], linestyle='dashdot', color='yellow', label='y = 3')

plt.xlabel("X label")
plt.ylabel("Y label")
plt.axis([1,4,0,9])
plt.legend(loc="upper right", ncol=4)

plt.show()
```

```python
# 튜플을 사용하여 선 종류 커스터마이즈
# 축 이름, 범위, 범례 설정
# tuple(offset, (on_off_seq))
# offset : 플롯의 간격 띄우기를 조정 (보통 0)

# 촘촘
plt.plot([1,2,3,4], [0.5, 1, 1.5, 2], linestyle=(0,(1,1)), color="C0", label="y = 1/2x")

# 넓은 간격
plt.plot([1,2,3,4], [1,2,3,4], linestyle=(0,(1,5)), color="red", label="y = x")

# 직선
plt.plot([1,2,3,4], [2,4,6,8], linestyle=(0,(5,1)), color="black", label="y = 2x")

# 직선, 점 반복
plt.plot([1,2,3,4], [3,3,3,3], linestyle=(0,(3,5,1,5)), color="yellow", label="y = 3")

plt.xlabel("X label")
plt.ylabel("Y label")
plt.axis([1,4,0,9])
plt.legend(loc="upper right", ncol=4)
plt.tight_layout() # 여백

plt.show()
```

# 영역을 지정하여 여러 개 그래프
- **subplot()** : 영역을 나눠 여러 그래프 시각화
- **plt.subplot(row, column, index)**로 공간 분할
- **tight_layout()** : 모서리의 여백 설정

Selena_파이썬 라이브러리 활용_STEP4_Matplotlib_13.jpg

```python
# linspace : 몇 등분할지 생각하면 디폴트는 50
# np.linsapce(0, 10) : 0부터 10까지 50등분한 결과를 배열로 반환
x1 = np.linspace(0, 10)
x1
```

```python
# linspace : 몇 등분할지 생각하면 디폴트는 50
# np.linsapce(0, 4) : 0부터 4까지 50등분한 결과를 배열로 반환
x2 = np.linspace(0, 4)
x2
```

```python
# y값은 역동적인 그래프를 위한 np.cosine 함수 이용
# np.pi 함수로 원주율 값 사용
y1 = np.cos(2*np.pi*x1)
y2 = np.cos(2*np.pi*x2)

# subplot1
# nrows=2, ncols=1, index=1
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title("1st Graph")

# subplot2
# nrows=2, ncols=1, index=2
plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.title("2nd Graph")

# tight_layout() 함수는 모서리와 서브플롯의 모서리 사이의 여백 설정
plt.tight_layout()

plt.show()
```

```python
# tight_layout() 이 없을 때
y1 = np.cos(2 *np.pi * x1)
y2 = np.cos(2 *np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title("1st Graph")

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.title("2nd Graph")

plt.show()
```

# 한 좌표평면 위에 여러 그래프 시각화
- plt.subplot**s**() : 한 좌표평면 위에 여러 그래프 시각화
- plt.subplot()의 디폴트 파라미터는 1
  - plt.subplots(nrows=1, ncols=1)
- plt.subplots()는 **figure와 axes**를 반환
  - **figure** : 전체 subplot
    - 서브플롯 안에 몇 개의 그래프가 있든 상관 없이, 그걸 담는 전체 사이즈
  - **axe** : 전체 중 낱낱개
    - 서브플롯 안에 2개의 그래프 a1, a2가 있다면, a1, a2
- **.twinx()** : ax1과 축을 공유하는 새로운 axes 객체 생성

```python
# x는 x축에 표시 될 연도
# y1, y2는 y값
x = ['2022', '2023', '2024']
y1 = np.array([1,7,14])
y2 = np.array([1,3,9])

# plt.subplots() : 여러 그래프를 한 번에 시각화, 객체 생성
fig, ax1 = plt.subplots()

# -s(solid line style + square marker), alpha(투명도)
ax1.plot(x, y1, '-s', color='g', markersize=7, linewidth=5, alpha=0.7)

# .txinx() : ax1과 축을 공유하는 새로운 axes 객체 생성
ax2 = ax1.twinx()
ax2.bar(x, y2, color='deeppink', width=0.7 ,alpha=0.7)

plt.show()
```
