```python
# 데이터 출처 : 기상자료개방포털 - 기후통계 -기온분석
# 원본 데이터의 공백을 제거하기 위한 처리
import csv
import pandas as pd
filename = '/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/myData.csv'
f = open(filename, encoding = 'cp949')
data = csv.reader(f)
next(data)
pfile = []

for row in data: # 한 줄씩
    row[0] = row[0].replace('\t', '') # \t를 제거해서
    pfile.append(row) # 빈 리스트에 넣기

print(pfile)
```

```python
# 판다스 DataFrame으로 변환
df = pd.DataFrame(pfile)
df.head(3)
```

```python
# 데이터형 확인
df.dtypes
```

```python
# 숫자형으로 변환
df[2] = pd.to_numeric(df[2], errors = 'coerce')
df[3] = pd.to_numeric(df[3], errors = 'coerce')
df[4] = pd.to_numeric(df[4], errors = 'coerce')
print(df.dtypes)
```

```python
# 열 이름 붙이기
df.columns = ['날짜', '지점', '평균기온', '최저기온', '최고기온']
df.columns
```

```python
# 날짜를 인덱스로 하여, 태어난 날짜의 특정 컬럼 선택
df.index = df['날짜']
df.loc['2002-02-14',['최저기온','최고기온']]
```

```python
# '평균기온' 칼럼에서 20이 넘는 행 추출
over20 = df['평균기온']>20
df[over20]
```
