```python
# 형태소 토큰화 - okt분석기 생성
from konlpy.tag import Okt
okt = Okt()
```

```python
# morphs() : 형태소 추출
token = okt.morphs('폐허가 된다 할지라도')
print(token)
```

```python
# pos() : 토큰과 품사정보를 튜플쌍으로 출력
sentence_tag = okt.pos('폐허가 된다 할지라도')
print(sentence_tag)
```

```python
# pos('문자열', join=True) : 문자열로 반환
sentence_tag = okt.pos('폐허가 된다 할지라도', join=True)
print(sentence_tag)
```

```python
# phrases() : 어구를 추출하여 반환
sentence = okt.phrases('한글 자연어 처리는 재밌다')
sentence
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import collections
filename='/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/1006실습/ratings_train(short).txt'
df = pd.read_table(filename)
df
```

```python
# 중복, 결측치 확인
print(df['id'].nunique())
print(df.isnull().sum())
```

```python
# 중복 제거, 결측치 확인
df = df.dropna(how='any')
print(df.isnull().values.any())
```

```python
# 불용어 제거, 결측치 확인
df['document'] = df['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣]","")
```

```python
# 중복 제거 및 결측치 확인
temp_list=[]
for sentence in df['document']:
    s_list = okt.pos(sentence)
    for word, tag in s_list:
        if tag in ['Noun', 'Adjective']:
            temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(50)
tag
```

```python
# 워드 클라우드 시각화
wc = WordCloud(font_path=font_path, background_color='skyblue', max_font_size=60)
cloud=wc.generate_from_frequencies(dict(tag))
plt.figure(figsize=(10,8))
plt.imshow(cloud)
```

```python
# 불용어 처리
list = []
stopword = ['점', '정말', '왜', '말', '그', '없다', '정도', '걸', '뭐', '이건', '영화', '완전', '좀', '있는', '거', '나', '이','볼','입니다','것','이런','더','수','때']
for sentence in df['document']:
    s_list = okt.pos(sentence)
    for word, tag in s_list:
        if word not in stopword:
            if tag in ['Noun', 'Adjective']:
                temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(50)
tag
```

```python
# 필요없는 단어 없애고 다시 워드 클라우드 시각화
wc = WordCloud(font_path=font_path, background_color='white', max_font_size=60)
cloud=wc.generate_from_frequencies(dict(tag))
plt.figure(figsize=(10,8))
plt.imshow(cloud)
```
