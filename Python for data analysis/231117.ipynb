# 과제 : 5단원 연습문제

```python
# 5단원 연습문제 1
# 과제
# 개방자료개방포털, 서울지역, 2002년 1~12월까지 자료를 다운 받아 판다스를 이용하여 읽고 연습문제 1-1 ~ 1-4까지 코드 작성
# 파일명 : 이름학과_과제2.py & 링크주소 & 사용한 csv파일(파일명 = mydata.csv) & 코랩결과화면캡처(이름학과_과제2결과화면.pdf -- ppt에 장마다 붙여넣기해서 pdf로 저장하기)

import csv
import pandas as pd
filename = '/content/drive/MyDrive/Colab Notebooks/2023-2 언분프/my_data.csv'
f = open(filename, encoding = 'cp949')
data = csv.reader(f)
next(data)
pfile = []

for row in data: # 한 줄씩
    row[0] = row[0].replace('\t', '') # \t를 제거해서
    pfile.append(row) # 빈 리스트에 넣기

print(pfile)
```

```python
df = pd.DataFrame(pfile) # 판다스 DataFrame으로 변환
df.head(3)
```

```python
df.dtypes # 데이터형 확인했더니 전부 문자열.,,
```

```python
df[2] = pd.to_numeric(df[2], errors = 'coerce') # 숫자형으로 변환
df[3] = pd.to_numeric(df[3], errors = 'coerce')
df[4] = pd.to_numeric(df[4], errors = 'coerce')
print(df.dtypes)
```
